```python
import os, glob
import pandas as pd
```

```python
path = 'E:\csv_clouds_do_polaczenia\dziesiec'
```

```python
all_files = glob.glob(os.path.join(path,"*.csv"))
```

```python
all_df = []
for f in all_files:
    df = pd.read_csv(f,sep=',')
    df['file'] = f.split('/')[-1]
    all_df.append(df)
    
print (all_df)
```

```python
merged_df = pd.concat(all_df, ignore_index = True, sort = True)
merged_df
```

```python
merged_df.to_csv(r'E:\cloud_10_OT_ETM_TM.csv')
```
