```python
import os, glob
import pandas as pd
```

```python
path = 'E:\csv_clouds_do_polaczenia\zero'
```

```python
#file1= 'E:\csv_clouds_do_polaczenia\zero\master_OT_cloud_0.csv'
#file2 = 'E:\csv_clouds_do_polaczenia\zero\master_ETM_cloud_0.csv'
#file3 = 'E:\csv_clouds_do_polaczenia\zero\master_TM_cloud_0.csv'
```

```python
all_files = glob.glob(os.path.join(path,"*.csv"))
#all_files = glob.glob(os.path.join(path,file1,file2))
```

```python
all_df = []
for f in all_files:
    df = pd.read_csv(f,sep=',')
    df['file'] = f.split('/')[-1]
    all_df.append(df)
    
print (all_df)
```

```python
merged_df = pd.concat(all_df, ignore_index = True, sort = True)
merged_df
```

```python
merged_df.to_csv(r'E:\cloud_0_OT_ETM_TM.csv')
```
