```python
import pandas as pd

df_landsat_masterdata = pd.read_csv('E:\metadata\LANDSAT_OT_C2_L2.csv')
```

```python
df_table_PR = pd.read_csv('E:\sceny_path_row.csv')
```

```python
print(df_table_PR)
```

```python
print(df_landsat_masterdata[['WRS Path','WRS Row']])
```

```python
path_row_dict = {}
master_dataframe = pd.DataFrame(columns = df_landsat_masterdata.columns)
print(len(master_dataframe))


for row in df_table_PR.itertuples():
    path_row_dict[row[0]] = (row[1], row[2])
    
for key, elem in path_row_dict.items():
    print("working on... " + str(elem))
    #master_dataframe.append((df_landsat_masterdata[(df_landsat_masterdata["WRS Path"]==elem[0]) & (df_landsat_masterdata["WRS Row"]==elem[1])]))
    df_temp = ((df_landsat_masterdata[(df_landsat_masterdata["WRS Path"]==elem[0]) & (df_landsat_masterdata["WRS Row"]==elem[1])]))
    print(len(df_temp))
    master_dataframe = master_dataframe.append(df_temp)
    print(len(master_dataframe))
    
#print(master_dataframe)
```

```python
print(master_dataframe)
```

```python
master_dataframe.to_csv(r'E:\master.csv')
```
