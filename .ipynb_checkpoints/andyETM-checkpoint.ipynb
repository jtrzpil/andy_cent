```python
import pandas as pd
```

```python
df_landsat_masterdata_ETM = pd.read_csv('E:\metadata\LANDSAT_ETM_C2_L2.csv')
```

```python
print(df_landsat_masterdata_ETM)
```

```python
df_table_PR = pd.read_csv('E:\sceny_path_row.csv')
```

```python
path_row_dict_ETM = {}
master_dataframe_ETM = pd.DataFrame(columns = df_landsat_masterdata_ETM.columns)
print(len(master_dataframe_ETM))


for row in df_table_PR.itertuples():
    path_row_dict_ETM[row[0]] = (row[1], row[2])
    
for key, elem in path_row_dict_ETM.items():
    print("working on... " + str(elem))
    df_temp_ETM = ((df_landsat_masterdata_ETM[(df_landsat_masterdata_ETM["WRS Path"]==elem[0]) & (df_landsat_masterdata_ETM["WRS Row"]==elem[1])]))
    print(len(df_temp_ETM))
    master_dataframe_ETM = master_dataframe_ETM.append(df_temp_ETM)
    print(len(master_dataframe_ETM))
    
```
