```python
path_row_dict_TM = {}
master_dataframe_TM = pd.DataFrame(columns = df_landsat_masterdata_TM.columns)
print(len(master_dataframe_TM))


for row in df_table_PR.itertuples():
    path_row_dict_TM[row[0]] = (row[1], row[2])
    
for key, elem in path_row_dict_TM.items():
    print("working on... " + str(elem))
    df_temp_TM = ((df_landsat_masterdata_TM[(df_landsat_masterdata_TM["WRS Path"]==elem[0]) & (df_landsat_masterdata_TM["WRS Row"]==elem[1])]))
    print(len(df_temp_TM))
    master_dataframe_TM = master_dataframe_TM.append(df_temp_TM)
    print(len(master_dataframe_TM))
    
```
